```python
import numpy as np
import os
import pandas as pd
from scipy import stats
import matplotlib.pyplot as plt
import plotly.graph_objects as go
import tarfile
import zipfile
from  sklearn.linear_model import LogisticRegression
import random
import math
import sounddevice as sd
import matplotlib.pyplot as plt
```

```python
train_data = pd.read_csv('../data/trainAudio.csv')
x_train = train_data['TARGET']
y_train = train_data['TARGET']
```

```python
test_data = pd.read_csv('../data/test.csv')
x_test = test_data['TARGET']
y_test = test_data['TARGET']
```

```python
def y_predict(a, b, c, d, x):
    return a * math.cos(b * x + c) + d
```

```python
lr = 0.05 
n = len(x_test)
batch = 100
epochs = 5000  
rsl = []
a = random.random()
b = random.random()
c = random.random()
d = 0

for i in range(epochs):
    a_gradiente = 0
    b_gradiente = 0
    c_gradiente = 0
    d_gradiente = 0
    e = 0
    for m in range(batch):
        ix = int(random.uniform(0, n))
        error = y_test[ix] - y_predict(a, b, c, d, x_test[ix])
        e += error ** 2
        
        # Calcular gradientes parciales
        a_gradiente += -2 * math.cos(b * x_test[ix] + c) * error
        b_gradiente += -2 * x_test[ix] * math.sin(b * x_test[ix] + c) * error
        c_gradiente += -2 * math.sin(b * x_test[ix] + c) * error
        d_gradiente += -2 * error

    a = a - lr * a_gradiente / batch
    b = b - lr * b_gradiente / batch
    c = c - lr * c_gradiente / batch

    if i in [599, 1799, 2500,3999,4999]:
        # Realizar las predicciones con los parámetros ajustados
        predicciones = [y_predict(a, b, c, d, x_value) for x_value in x_train]
    
        # Graficar la onda de sonido de los datos reales y las predicciones
        plt.figure(figsize=(16, 12))
        plt.plot(x_train, predicciones, label=f'Predicciones (Época {i+1})', color='r')
        plt.xlabel('Tiempo')
        plt.ylabel('Amplitud')
        plt.title('Onda de Sonido')
        plt.legend()
        plt.grid(True)
        plt.show()

    e = e / batch
    rsl.append([a, b, c, d, e])
    print(f"Iteración {i+1}: Error={e}, Periodo={b}, Amplitud={a}, Constante={c}")
```

```python
predictions = [y_predict(a, b, c, d, x_value) for x_value in x_test]

resultado_df = pd.DataFrame({'example_id': range(13000, 13000 + len(predictions)), 'TARGET': predictions})

resultado_df.to_csv('predicciones3.csv', index=False)
```
